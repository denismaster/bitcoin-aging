Let's load some bitcoin datasets to understand it's processes in dynamic:

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

from matplotlib.pylab import rcParams
%matplotlib inline  

# function to load csv file and parse provided column as date
def load_file(filename, date_col):
    dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d %H:%M:%S')
    data = pd.read_csv(filename,index_col=date_col, parse_dates=[0],date_parser=dateparse)
    return data

hash_rate_data = load_file('hash-rate.csv', 'Date');
hash_rate_series = hash_rate_data['Data']

difficulty_data = load_file('difficulty.csv', 'Date');
difficulty_series = difficulty_data['Data']

transactions_count_data = load_file('n-transactions.csv', 'Date');
transactions_count_series = transactions_count_data['Data']

fee_percent_data = load_file('cost-per-transaction-percent.csv', 'Date');
fee_percent_series = fee_percent_data['Data']

def plot_chart(data, label, title):
    plt.plot(data, label=label)
    plt.legend(loc='best')
    plt.title(title)
    plt.show()

plot_chart(hash_rate_series, "Hash-rate", "Hash-rate")


plt.plot(difficulty_series, label="Difficulty")
plt.legend(loc='best')
plt.title('Difficulty')
plt.show()

plt.plot(transactions_count_series, label="Transactions count")
plt.legend(loc='best')
plt.title('Transactions count')
plt.show()

plt.plot(fee_percent_series, label="Transaction fee(percent)")
plt.legend(loc='best')
plt.title('Transaction fee(percent)')
plt.show()
```
