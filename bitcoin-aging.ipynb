Let's load some bitcoin datasets to understand it's processes in dynamic:

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

from matplotlib.pylab import rcParams
%matplotlib inline  

# function to load csv file and parse provided column as date
def load_file(filename, date_col):
    dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m-%d %H:%M:%S')
    data = pd.read_csv(filename,index_col=date_col, parse_dates=[0],date_parser=dateparse)
    return data

hash_rate_data = load_file('hash-rate.csv', 'Date');
hash_rate_series = hash_rate_data['Data']

difficulty_data = load_file('difficulty.csv', 'Date');
difficulty_series = difficulty_data['Data']

tx_count_data = load_file('n-transactions.csv', 'Date');
tx_count_series = tx_count_data['Data']

tx_fee_data = load_file('cost-per-transaction-percent.csv', 'Date');
tx_fee_series = tx_fee_data['Data']
```

Firstly we need to display charts of our datasets:

```python
def plot_chart(data, label, title):
    plt.plot(data, label=label)
    plt.legend(loc='best')
    plt.title(title)
    plt.show()

plot_chart(hash_rate_series, "Hash-rate", "Hash-rate")
plot_chart(difficulty_series, "Difficulty", "Bitcoin difficulty")
plot_chart(tx_count_series, "Tx count", "Transaction count")
plot_chart(tx_fee_series, "Tx fee", "Transaction fee(percentage)")
```
